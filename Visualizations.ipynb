```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
PC45 = pd.read_csv("results/PC45.csv")
PC35 = pd.read_csv("results/PC35.csv")
KC15 = pd.read_csv("results/KC15.csv")
```

```python
PC45 = PC45.dropna()
PC35 = PC35.dropna()
KC15 = KC15.dropna()
```

```python
PC45.index = [i for i in range(PC45.shape[0])]
PC35.index = [i for i in range(PC35.shape[0])]
KC15.index = [i for i in range(KC15.shape[0])]
```

```python
fig,ax = plt.subplots(figsize = (7,5))
dfs = [PC35,PC45,KC15]
x = ["PC3","PC4","KC1"]
YSVM = [PC35.iloc[13]['Average Recall'],PC45.iloc[12]['Average Recall'],KC15.iloc[12]['Average Recall']]
YSVMe = [PC35.iloc[13]['Standard Deviation'],PC45.iloc[12]['Standard Deviation'],KC15.iloc[12]['Standard Deviation']]
ax.errorbar(x,YSVM,YSVMe,label = "Oversampled Data (5 Fold)",color = "blue")
YSVM = [0.19,0.26,0.24]
YSVMe = [0.08,0.07,0.035]
ax.errorbar(x,YSVM,YSVMe,label = "Original Data (5 Fold)",color = "red")
ax.set_title('SVM')
ax.set_ylabel("Recall")
ax.set_xlabel("Datasets")
ax.legend(loc='best')
ax.set_ylim(0,1)
fig.savefig('svm.png', dpi=125)
```
